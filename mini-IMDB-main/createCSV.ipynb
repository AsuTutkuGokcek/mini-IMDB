```python
# IMPORTS ARE HERE
import pandas as pd
import numpy as np
```

```python
title_akas_df = pd.read_csv('title.akas.tsv.gz', compression='gzip', header=0, sep='\t', quotechar='"', error_bad_lines=False)
episode_df = pd.read_csv('title.episode.tsv.gz', compression='gzip', header=0, sep='\t', quotechar='"', error_bad_lines=False)
basics_df = pd.read_csv('title.basics.tsv.gz', compression='gzip', header=0, sep='\t', quotechar='"', error_bad_lines=False)
ratings_df = pd.read_csv('title.ratings.tsv.gz', compression='gzip', header=0, sep='\t', quotechar='"', error_bad_lines=False)
crew_df = pd.read_csv('title.crew.tsv.gz',compression='gzip', header=0, sep='\t', quotechar='"', error_bad_lines=False)
principals_df = pd.read_csv('title.principals.tsv.gz', compression='gzip', header=0, sep='\t', quotechar='"', error_bad_lines=False)
name_basic_df = pd.read_csv('name.basics.tsv.gz', compression='gzip', header=0, sep='\t', quotechar='"', error_bad_lines=False)
```

```python
# create DIRECTOR CSV
name_basic_db_df = name_basic_df[~name_basic_df['primaryProfession'].isna()]
director_db_df = name_basic_db_df[name_basic_db_df['primaryProfession'].str.contains('director')]
director_db_df = director_db_df[['nconst', 'primaryName', 'birthYear', 'deathYear']]
director_db_df = director_db_df[~(director_db_df['birthYear'] == '\\N')]
director_db_df['deathYear'] = director_db_df['deathYear'].replace('\\N', np.nan)
director_db_df.to_csv('director.csv', index=False)
director_db_df.head(60)
```

```python
# create TV-SERIES CSV
series_db_df = basics_df[basics_df['titleType'] == 'tvSeries']
series_db_df = series_db_df[~(series_db_df['genres'] == '\\N')]
series_db_df = series_db_df[~(series_db_df['startYear'] == '\\N')]
series_db_df = pd.merge(series_db_df, ratings_df, on='tconst', how='inner')
series_db_df = series_db_df.drop_duplicates()
no_episodes = episode_db_df['parentTconst'].value_counts().to_frame().reset_index()
no_episodes.columns = ['tconst', 'no_episodes']
series_db_df = pd.merge(series_db_df, no_episodes, on='tconst', how='inner')
series_db_df = series_db_df.drop_duplicates()
series_db_df = series_db_df[['tconst', 'primaryTitle', 'averageRating', 'numVotes', 'startYear', 'no_episodes']]
series_db_df = series_db_df[series_db_df['primaryTitle'].str.len() < 100]
series_db_df.to_csv('tv_series.csv', index=False)
series_db_df
```

```python
# create Actor CSV
name_basic_db_df = name_basic_df[~name_basic_df['primaryProfession'].isna()]
actors_db_df = name_basic_db_df[(name_basic_db_df['primaryProfession'].str.contains('actor')) | (name_basic_db_df['primaryProfession'].str.contains('actress'))]
actors_db_df = actors_db_df[~(actors_db_df['birthYear'] == '\\N')]
actors_db_df = actors_db_df[['nconst', 'primaryName', 'birthYear', 'deathYear']]
actors_db_df = actors_db_df[actors_db_df['primaryName'].str.len() < 100]
actors_db_df['deathYear'] = actors_db_df['deathYear'].replace('\\N', np.nan)
actors_db_df.to_csv('actors.csv', index=False)
actors_db_df
```

```python
# create plays_in_movies CSV
plays_in_movie_db_df = principals_df[principals_df['tconst'].isin(movie_db_df['tconst'].unique())]
plays_in_movie_db_df = plays_in_movie_db_df[plays_in_movie_db_df['category'].isin(['actor','actress'])]
plays_in_movie_db_df = plays_in_movie_db_df[['tconst', 'nconst']]
plays_in_movie_db_df = plays_in_movie_db_df.drop_duplicates()
plays_in_movie_db_df = plays_in_movie_db_df[plays_in_movie_db_df['nconst'].isin(actors_db_df['nconst'].unique())]
plays_in_movie_db_df.to_csv('plays_in_movies.csv', index=False)
plays_in_movie_db_df
```

```python
# create plays_in_series CSV
plays_in_series_db_df = principals_df[principals_df['tconst'].isin(series_db_df['tconst'].unique())]
plays_in_series_db_df = plays_in_series_db_df[plays_in_series_db_df['category'].isin(['actor','actress'])]
plays_in_series_db_df = plays_in_series_db_df[['tconst', 'nconst']]
plays_in_series_db_df = plays_in_series_db_df.drop_duplicates()
plays_in_series_db_df.to_csv('plays_in_series.csv', index=False)
plays_in_series_db_df
```

```python
# create genre_movies CSV
movie_list_db_df = basics_df[basics_df['titleType'] == 'movie']
movie_genre_db_df = movie_list_db_df.join(movie_list_db_df['genres'].str.split(',', expand=True).rename(columns={0:'Genre1', 1:'Genre2', 2:'Genre3'}))
movie_genre_db_df = movie_genre_db_df[['tconst', 'Genre1', 'Genre2', 'Genre3']]
movie_genre_db_df = movie_genre_db_df[~(movie_genre_db_df['Genre1'] == '\\N')]
pd.melt(movie_genre_db_df, id_vars='tconst', value_vars=['Genre1','Genre2','Genre3'], value_name='Genre')
movie_genre_db_df = movie_genre_db_df.melt('tconst', value_name='Genre').drop('variable', 1)
movie_genre_db_df = movie_genre_db_df[~movie_genre_db_df['Genre'].isna()]
movie_genre_db_df = movie_genre_db_df[movie_genre_db_df['tconst'].isin(movie_db_df['tconst'].unique())]
movie_genre_db_df.to_csv('genre_movies.csv', index=False)
movie_genre_db_df
```
